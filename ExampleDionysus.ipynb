Ejemplo calcular lower_star filtration y diagrama en dionysus: https://mrzv.org/software/dionysus2/tutorial/lower-star.html

```python
import dionysus as d
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.array([1,2,3])
y = np.array([[0],[2],[1]],dtype=np.float32)
print(y.shape)
plt.plot(t,y,'-o')
plt.title(f"Signal")
plt.xlim(0,4)
plt.ylim(-1,4)
plt.show()
```

```python
f_lower_star = d.fill_freudenthal(y)
f_lower_star
```

```python
p = d.homology_persistence(f_lower_star)
print(p)
dgms = d.init_diagrams(p, f_lower_star)
dgms[0]
```

```python
for i in range(2):
    print(dgms[0][i])
```

```python
d.plot.plot_diagram(dgms[0])
```
